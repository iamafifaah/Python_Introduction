# Continue, Pass and Break

# 1. Pass
Pass tidak akan dijalankan dalam suatu program

```python
angka = 0

while angka < 5:
    angka += 1
    if angka == 3:
        print("okeyyyy")

    print(angka)
```

```python
## Jika ditambah pass, maka program tidak akan dieksekusi

angka = 0

while angka < 5:
    angka += 1
    if angka == 3:
        pass
        #print("okeyyyy")
        

    print(angka)
```

# 2. Continue

```python
angka = 0

while angka < 10:
    angka += 1
    print(f"angka sekarang -> {angka}")

    if angka == 3:
        print("nice!!!") ### statement if akan dijalankan sebagai aksi ke-1 pada iterasi jika i == 3 (sama dengan)

    print("okeyyyy")

print("finish")
```

```python
angka = 0

while angka < 10:
    angka += 1
    print(f"angka sekarang -> {angka}")

    if angka == 3:
        print("nice!!!") ### statement if akan dijalankan sebagai aksi ke-1 pada iterasi jika i == 3 (sama dengan)
        continue ### continue membuat program langsung kembali ke awal dan melewati semua program setelahnya
    print("okeyyyy") # program yang terlewati karena ada continue

print("finish")
```

# 3. Break

```python
angka = 0

while angka < 10:
    angka += 1
    print(angka)

print("stop")

# contoh while yang bisa berhenti, karena ketika kondisi sudah terpenuhi sampai batasan yang ditentukan, program akan berhenti
```

```python
angka = 0

while angka < 30:
    angka += 1
    print(angka)
    if angka == 10:
        print("yess")
    if angka == 20:
        break
    print("ookee")

print("selesai!!!")

### program while dengan ada statement break
```

```python
angka = 0

while angka < 30:
    angka += 1
    print(angka)
    if angka == 10:
        print("yess")
        continue
    if angka == 20:
        print("asdfghjkl")
        break
    print("ookee")

print("selesai!!!")

### program while dengan ada statement continue, dan break
## ketika iterasi angka ==10 dia akan mengabaikan aksi selanjutnya dan lanjut ke iterasi angka ==11
```
